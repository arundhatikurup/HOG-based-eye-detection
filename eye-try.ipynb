```python
import os
import numpy 
import cv2
from matplotlib import pyplot as plt
from PIL import Image
from skimage import exposure
from skimage import feature

path="/home/arundhati/cv/dlib_project/eye"
image_paths = [os.path.join(path, f) for f in os.listdir(path)]

data=[]
for image_path in image_paths:
    print image_path
    image_pil = Image.open(image_path).convert('L')
    image = numpy.array(image_pil, 'uint8')
    H = feature.hog(image, orientations=9, pixels_per_cell=(10, 10),cells_per_block=(2,2), transform_sqrt=True)
    a=numpy.transpose(H)
    data.append(a)
print data

fig, ax = plt.subplots()


for i in range(0,len(data),1):
    plt.xlabel('bins')
    plt.ylabel('magnitude')
    plt.title('Histogram of OG')
    plt.hist(data[i],orientation='vertical')
    plt.show()
    
```

```python
def hog(image_path):    
    image_pil = Image.open(image_path).convert('L')
    image = numpy.array(image_pil, 'uint8')
    cv2.imshow("image",image)
    H = feature.hog(image, orientations=9, pixels_per_cell=(10, 10),cells_per_block=(2,2), transform_sqrt=True)
    rowcount=H.shape[0]/9
    H.resize(rowcount,9)
    print H.shape
    plt.xlabel('bins')
    plt.ylabel('mag')
    plt.title('Histogram of OG')
    a=np.sum(H, axis=0)
    plt.hist(a,[1,2,3,4,5,6,7,8,9])
    plt.show()



linka="/home/arundhati/cv/dlib_project/Alpha/ab.jpg"
linkaa="/home/arundhati/cv/dlib_project/Alpha/aab.jpg"
hog(linka)
hog(linkaa)

```
